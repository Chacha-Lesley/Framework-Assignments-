```python
# CORD-19 Dataset Analysis - Complete Jupyter Notebook
# This notebook provides a step-by-step analysis of the CORD-19 dataset

# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
import re
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')

# Set up plotting style
plt.style.use('seaborn-v0_8')
plt.rcParams['figure.figsize'] = (12, 8)
sns.set_palette("husl")

print("üìö CORD-19 Dataset Analysis Notebook")
print("=" * 50)
print("This notebook will guide you through analyzing the CORD-19 research dataset")
print("=" * 50)

# ============================================================================
# PART 1: DATA LOADING AND BASIC EXPLORATION
# ============================================================================

print("\nüîç PART 1: DATA LOADING AND BASIC EXPLORATION")
print("-" * 50)

# Load the dataset
def load_cord19_data(file_path='metadata.csv', sample_size=None, verbose=True):
    """
    Load the CORD-19 metadata dataset
    
    Parameters:
    - file_path (str): Path to the metadata.csv file
    - sample_size (int): Number of rows to load (None for all)
    - verbose (bool): Print loading information
    
    Returns:
    - pandas.DataFrame: The loaded dataset
    """
    try:
        if verbose:
            print(f"üìÇ Loading data from {file_path}...")
            if sample_size:
                print(f"   ‚Üí Loading sample of {sample_size:,} rows")
            else:
                print("   ‚Üí Loading full dataset")
        
        # Load the data
        if sample_size:
            df = pd.read_csv(file_path, nrows=sample_size)
        else:
            df = pd.read_csv(file_path)
        
        if verbose:
            print(f"‚úÖ Data loaded successfully!")
            print(f"   ‚Üí Shape: {df.shape}")
            print(f"   ‚Üí Memory usage: {df.memory_usage(deep=True).sum() / 1024**2:.1f} MB")
        
        return df
    
    except FileNotFoundError:
        print("‚ùå Error: File not found!")
        print("Please download metadata.csv from:")
        print("https://www.kaggle.com/allen-institute-for-ai/CORD-19-research-challenge")
        return None
    
    except Exception as e:
        print(f"‚ùå Error loading data: {e}")
        return None

# Load the data (adjust the path and sample size as needed)
df = load_cord19_data('metadata.csv', sample_size=10000)

if df is not None:
    # Basic data exploration
    print("\nüìä BASIC DATA INFORMATION")
    print("-" * 30)
    print(f"Dataset dimensions: {df.shape[0]:,} rows √ó {df.shape[1]} columns")
    print(f"Date range: {df.shape}")
    
    # Display first few rows
    print("\nüìã FIRST 5 ROWS:")
    display(df.head())
    
    # Column information
    print(f"\nüìù COLUMN INFORMATION:")
    print(f"Total columns: {len(df.columns)}")
    for i, col in enumerate(df.columns, 1):
        print(f"{i:2d}. {col}")
    
    # Data types
    print(f"\nüî¢ DATA TYPES:")
    print(df.dtypes)
    
    # Basic statistics
    print(f"\nüìà BASIC STATISTICS:")
    display(df.describe())
    
    # Missing values analysis
    print(f"\n‚ùì MISSING VALUES ANALYSIS:")
    missing_data = pd.DataFrame({
        'Column': df.columns,
        'Missing_Count': df.isnull().sum(),
        'Missing_Percentage': (df.isnull().sum() / len(df)) * 100
    }).sort_values('Missing_Percentage', ascending=False)
    
    display(missing_data)
    
    # Plot missing values
    plt.figure(figsize=(12, 6))
    missing_data_plot = missing_data[missing_data['Missing_Percentage'] > 0]
    
    if len(missing_data_plot) > 0:
        plt.barh(missing_data_plot['Column'], missing_data_plot['Missing_Percentage'])
        plt.xlabel('Missing Percentage (%)')
        plt.title('Missing Values by Column')
        plt.tight_layout()
        plt.show()
    else:
        print("‚úÖ No missing values found in the dataset!")

# ============================================================================
# PART 2: DATA CLEANING AND PREPARATION
# ============================================================================

print("\nüßπ PART 2: DATA CLEANING AND PREPARATION")
print("-" * 50)

def clean_cord19_data(df, verbose=True):
    """
    Clean and prepare the CORD-19 dataset
    
    Parameters:
    - df (pandas.DataFrame): Raw dataset
    - verbose (bool): Print cleaning steps
    
    Returns:
    - pandas.DataFrame: Cleaned dataset
    """
    if df is None:
        return None
    
    if verbose:
        print("üßΩ Starting data cleaning process...")
    
    # Create a copy for cleaning
    df_clean = df.copy()
    original_rows = len(df_clean)
    
    # 1. Handle missing values in key columns
    if verbose:
        print("\n1Ô∏è‚É£ Handling missing values...")
    
    # Fill missing titles
    if 'title' in df_clean.columns:
        missing_titles = df_clean['title'].isnull().sum()
        df_clean['title'] = df_clean['title'].fillna('Unknown Title')
        if verbose and missing_titles > 0:
            print(f"   ‚Üí Filled {missing_titles:,} missing titles")
    
    # Fill missing abstracts
    if 'abstract' in df_clean.columns:
        missing_abstracts = df_clean['abstract'].isnull().sum()
        df_clean['abstract'] = df_clean['abstract'].fillna('No abstract available')
        if verbose and missing_abstracts > 0:
            print(f"   ‚Üí Filled {missing_abstracts:,} missing abstracts")
    
    # Fill missing journal names
    if 'journal' in df_clean.columns:
        missing_journals = df_clean['journal'].isnull().sum()
        df_clean['journal'] = df_clean['journal'].fillna('Unknown Journal')
        if verbose and missing_journals > 0:
            print(f"   ‚Üí Filled {missing_journals:,} missing journal names")
    
    # 2. Process date columns
    if verbose:
        print("\n2Ô∏è‚É£ Processing date columns...")
    
    date_columns = ['publish_time', 'publish_date']
    for col in date_columns:
        if col in df_clean.columns:
            # Convert to datetime
            original_count = df_clean[col].notna().sum()
            df_clean[col] = pd.to_datetime(df_clean[col], errors='coerce')
            converted_count = df_clean[col].notna().sum()
            
            # Create year column
            year_col = col.replace('_time', '_year').replace('_date', '_year')
            df_clean[year_col] = df_clean[col].dt.year
            
            if verbose:
                print(f"   ‚Üí Processed {col}: {converted_count:,}/{original_count:,} valid dates")
                print(f"   ‚Üí Created {year_col} column")
    
    # 3. Create derived columns
    if verbose:
        print("\n3Ô∏è‚É£ Creating derived columns...")
    
    # Title word count
    if 'title' in df_clean.columns:
        df_clean['title_word_count'] = df_clean['title'].apply(
            lambda x: len(str(x).split()) if pd.notna(x) else 0
        )
        if verbose:
            avg_title_length = df_clean['title_word_count'].mean()
            print(f"   ‚Üí Created title_word_count (avg: {avg_title_length:.1f} words)")
    
    # Abstract word count
    if 'abstract' in df_clean.columns:
        df_clean['abstract_word_count'] = df_clean['abstract'].apply(
            lambda x: len(str(x).split()) if pd.notna(x) and x != 'No abstract available' else 0
        )
        if verbose:
            avg_abstract_length = df_clean['abstract_word_count'].mean()
            print(f"   ‚Üí Created abstract_word_count (avg: {avg_abstract_length:.1f} words)")
    
    # 4. Data quality filters
    if verbose:
        print("\n4Ô∏è‚É£ Applying data quality filters...")
    
    # Remove rows with placeholder titles (if we want to)
    # df_clean = df_clean[df_clean['title'] != 'Unknown Title']
    
    # Filter reasonable publication years (COVID-19 research context)
    if 'publish_year' in df_clean.columns:
        reasonable_years = df_clean['publish_year'].between(1900, 2025, na=False)
        df_clean = df_clean[reasonable_years | df_clean['publish_year'].isna()]
        filtered_years = len(df_clean)
        if verbose:
            removed_years = original_rows - filtered_years
            print(f"   ‚Üí Filtered unreasonable years: removed {removed_years:,} rows")
    
    if verbose:
        final_rows = len(df_clean)
        print(f"\n‚úÖ Data cleaning complete!")
        print(f"   ‚Üí Final dataset: {final_rows:,} rows")
        print(f"   ‚Üí Rows removed: {original_rows - final_rows:,}")
        print(f"   ‚Üí Data retention: {(final_rows/original_rows)*100:.1f}%")
    
    return df_clean

# Clean the data
if df is not None:
    df_clean = clean_cord19_data(df)
    
    # Show before/after comparison
    if df_clean is not None:
        print(f"\nüìä BEFORE/AFTER COMPARISON:")
        comparison = pd.DataFrame({
            'Metric': ['Rows', 'Columns', 'Memory (MB)'],
            'Original': [
                f"{df.shape[0]:,}",
                f"{df.shape[1]:,}",
                f"{df.memory_usage(deep=True).sum() / 1024**2:.1f}"
            ],
            'Cleaned': [
                f"{df_clean.shape[0]:,}",
                f"{df_clean.shape[1]:,}",
                f"{df_clean.memory_usage(deep=True).sum() / 1024**2:.1f}"
            ]
        })
        display(comparison)

# ============================================================================
# PART 3: DATA ANALYSIS AND VISUALIZATION
# ============================================================================

print("\nüìä PART 3: DATA ANALYSIS AND VISUALIZATION")
print("-" * 50)

if df_clean is not None:
    # Create a comprehensive analysis
    def create_publication_timeline(df):
        """Create publication timeline visualization"""
        if 'publish_year' in df.columns:
            print("üìÖ PUBLICATION TIMELINE ANALYSIS")
            
            # Get year counts
            year_counts = df['publish_year'].value_counts().sort_index()
            
            # Filter to reasonable range
            year_counts = year_counts[(year_counts.index >= 2000) & (year_counts.index <= 2023)]
            
            # Create the plot
            plt.figure(figsize=(14, 6))
            plt.subplot(1, 2, 1)
            plt.plot(year_counts.index, year_counts.values, marker='o', linewidth=2, markersize=6)
            plt.title('Publications by Year', fontsize=14, fontweight='bold')
            plt.xlabel('Year')
            plt.ylabel('Number of Publications')
            plt.grid(True, alpha=0.3)
            
            # Highlight COVID-19 period
            covid_years = year_counts[year_counts.index >= 2020]
            if len(covid_years) > 0:
                plt.axvspan(2020, year_counts.index.max(), alpha=0.2, color='red', 
                           label='COVID-19 Period')
                plt.legend()
            
            # Bar chart for recent years
            plt.subplot(1, 2, 2)
            recent_years = year_counts[year_counts.index >= 2018]
            plt.bar(recent_years.index, recent_years.values, color='skyblue', alpha=0.7)
            plt.title('Recent Publications (2018+)', fontsize=14, fontweight='bold')
            plt.xlabel('Year')
            plt.ylabel('Number of Publications')
            
            plt.tight_layout()
            plt.show()
            
            # Print insights
            peak_year = year_counts.idxmax()
            peak_count = year_counts.max()
            print(f"   üìà Peak publication year: {peak_year} ({peak_count:,} papers)")
            
            if len(covid_years) > 1:
                covid_growth = ((covid_years.iloc[-1] - covid_years.iloc[0]) / covid_years.iloc[0]) * 100
                print(f"   ü¶† COVID-19 period growth: {covid_growth:.1f}%")
        
        else:
            print("‚ùå Publication year data not available")
    
    def analyze_journals(df, top_n=15):
        """Analyze journal distribution"""
        if 'journal' in df.columns:
            print(f"\nüì∞ TOP {top_n} JOURNALS ANALYSIS")
            
            journal_counts = df['journal'].value_counts().head(top_n)
            
            # Create visualization
            plt.figure(figsize=(14, 8))
            
            # Horizontal bar chart
            plt.subplot(2, 1, 1)
            plt.barh(range(len(journal_counts)), journal_counts.values, color='lightgreen', alpha=0.7)
            plt.yticks(range(len(journal_counts)), 
                      [j[:50] + '...' if len(j) > 50 else j for j in journal_counts.index])
            plt.xlabel('Number of Papers')
            plt.title(f'Top {top_n} Publishing Journals', fontsize=14, fontweight='bold')
            plt.gca().invert_yaxis()
            
            # Pie chart for top 10
            plt.subplot(2, 1, 2)
            top_10_journals = journal_counts.head(10)
            plt.pie(top_10_journals.values, labels=top_10_journals.index, autopct='%1.1f%%')
            plt.title('Top 10 Journals Distribution', fontsize=14, fontweight='bold')
            
            plt.tight_layout()
            plt.show()
            
            # Print top journals
            print(f"   ü•á Top 5 Journals:")
            for i, (journal, count) in enumerate(journal_counts.head(5).items(), 1):
                percentage = (count / len(df)) * 100
                print(f"      {i}. {journal}: {count:,} papers ({percentage:.1f}%)")
        
        else:
            print("‚ùå Journal data not available")
    
    def analyze_text_patterns(df):
        """Analyze text patterns in titles and abstracts"""
        print(f"\nüìù TEXT ANALYSIS")
        
        # Word count distributions
        fig, axes = plt.subplots(2, 2, figsize=(15, 10))
        
        # Title word count distribution
        if 'title_word_count' in df.columns:
            axes[0, 0].hist(df['title_word_count'].dropna(), bins=30, 
                           color='skyblue', alpha=0.7, edgecolor='black')
            axes[0, 0].set_title('Title Word Count Distribution')
            axes[0, 0].set_xlabel('Number of Words')
            axes[0, 0].set_ylabel('Frequency')
            
            mean_title_words = df['title_word_count'].mean()
            axes[0, 0].axvline(mean_title_words, color='red', linestyle='--', 
                              label=f'Mean: {mean_title_words:.1f}')
            axes[0, 0].legend()
        
        # Abstract word count distribution
        if 'abstract_word_count' in df.columns:
            axes[0, 1].hist(df['abstract_word_count'].dropna(), bins=30, 
                           color='lightcoral', alpha=0.7, edgecolor='black')
            axes[0, 1].set_title('Abstract Word Count Distribution')
            axes[0, 1].set_xlabel('Number of Words')
            axes[0, 1].set_ylabel('Frequency')
            
            mean_abstract_words = df['abstract_word_count'].mean()
            axes[0, 1].axvline(mean_abstract_words, color='red', linestyle='--', 
                              label=f'Mean: {mean_abstract_words:.1f}')
            axes[0, 1].legend()
        
        # Common words in titles
        if 'title' in df.columns:
            common_words = get_common_words_from_text(df['title'], top_n=15)
            if common_words:
                words, counts = zip(*common_words[:10])
                axes[1, 0].barh(range(len(words)), counts, color='gold', alpha=0.7)
                axes[1, 0].set_yticks(range(len(words)))
                axes[1, 0].set_yticklabels(words)
                axes[1, 0].set_title('Top 10 Words in Titles')
                axes[1, 0].set_xlabel('Frequency')
                axes[1, 0].invert_yaxis()
        
        # Source distribution
        source_col = None
        for col in ['source_x', 'source']:
            if col in df.columns:
                source_col = col
                break
        
        if source_col:
            source_counts = df[source_col].value_counts().head(8)
            axes[1, 1].pie(source_counts.values, labels=source_counts.index, autopct='%1.1f%%')
            axes[1, 1].set_title('Distribution by Source')
        else:
            axes[1, 1].text(0.5, 0.5, 'Source data\nnot available', 
                           ha='center', va='center', fontsize=12)
            axes[1, 1].set_title('Source Distribution')
        
        plt.tight_layout()
        plt.show()
    
    def get_common_words_from_text(text_series, top_n=20):
        """Extract common words from a text series"""
        # Combine all text
        text = ' '.join(text_series.dropna().astype(str))
        
        # Define stop words
        stop_words = {
            'the', 'a', 'an', 'and', 'or', 'but', 'in', 'on', 'at', 'to', 'for', 'of', 'with', 'by',
            'is', 'are', 'was', 'were', 'be', 'been', 'being', 'have', 'has', 'had', 'do', 'does',
            'did', 'will', 'would', 'could', 'should', 'may', 'might', 'must', 'can', 'this', 'that',
            'these', 'those', 'i', 'you', 'he', 'she', 'it', 'we', 'they', 'me', 'him', 'her', 'us',
            'them', 'covid', 'coronavirus', 'sars', 'cov', 'pandemic', 'study', 'analysis', 'research',
            'paper', 'article', 'review', 'disease', 'virus', 'using', 'based', 'case', 'cases'
        }
        
        # Extract words
        words = re.findall(r'\b[a-zA-Z]+\b', text.lower())
        words = [word for word in words if word not in stop_words and len(word) > 2]
        
        # Count frequency
        word_counts = Counter(words).most_common(top_n)
        
        return word_counts
    
    def generate_summary_report(df):
        """Generate a comprehensive summary report"""
        print(f"\nüìã COMPREHENSIVE SUMMARY REPORT")
        print("=" * 50)
        
        # Dataset overview
        print(f"üìä DATASET OVERVIEW:")
        print(f"   ‚Ä¢ Total papers analyzed: {len(df):,}")
        print(f"   ‚Ä¢ Dataset columns: {df.shape[1]}")
        print(f"   ‚Ä¢ Memory usage: {df.memory_usage(deep=True).sum() / 1024**2:.1f} MB")
        
        # Time period analysis
        if 'publish_year' in df.columns:
            years = df['publish_year'].dropna()
            if len(years) > 0:
                print(f"\nüìÖ TIME PERIOD:")
                print(f"   ‚Ä¢ Year range: {int(years.min())} - {int(years.max())}")
                print(f"   ‚Ä¢ Span: {int(years.max() - years.min())} years")
                print(f"   ‚Ä¢ Most productive year: {int(years.mode().iloc[0])}")
        
        # Content analysis
        if 'journal' in df.columns:
            journals = df['journal'].value_counts()
            print(f"\nüì∞ CONTENT DISTRIBUTION:")
            print(f"   ‚Ä¢ Unique journals: {len(journals):,}")
            print(f"   ‚Ä¢ Top journal: {journals.index[0]} ({journals.iloc[0]:,} papers)")
        
        if 'title_word_count' in df.columns:
            avg_title_length = df['title_word_count'].mean()
            print(f"\nüìù TEXT CHARACTERISTICS:")
            print(f"   ‚Ä¢ Average title length: {avg_title_length:.1f} words")
            
        if 'abstract_word_count' in df.columns:
            avg_abstract_length = df['abstract_word_count'].mean()
            print(f"   ‚Ä¢ Average abstract length: {avg_abstract_length:.1f} words")
        
        # Common research topics
        if 'title' in df.columns:
            common_words = get_common_words_from_text(df['title'], top_n=10)
            if common_words:
                print(f"\nüîç TOP RESEARCH TOPICS (from titles):")
                for i, (word, count) in enumerate(common_words[:5], 1):
                    print(f"   {i}. {word.title()}: {count:,} mentions")
        
        # Data quality assessment
        print(f"\n‚úÖ DATA QUALITY:")
        missing_critical = 0
        critical_columns = ['title', 'journal', 'publish_time']
        for col in critical_columns:
            if col in df.columns:
                missing_pct = (df[col].isnull().sum() / len(df)) * 100
                if missing_pct > 10:
                    missing_critical += 1
                print(f"   ‚Ä¢ {col} completeness: {100 - missing_pct:.1f}%")
        
        if missing_critical == 0:
            print("   ‚Ä¢ Overall data quality: GOOD ‚úÖ")
        elif missing_critical <= 2:
            print("   ‚Ä¢ Overall data quality: FAIR ‚ö†Ô∏è")
        else:
            print("   ‚Ä¢ Overall data quality: POOR ‚ùå")
    
    # Run all analyses
    create_publication_timeline(df_clean)
    analyze_journals(df_clean, top_n=12)
    analyze_text_patterns(df_clean)
    generate_summary_report(df_clean)

else:
    print("‚ùå No data available for analysis")

# ============================================================================
# PART 4: ADDITIONAL INSIGHTS AND RECOMMENDATIONS
# ============================================================================

print(f"\nüéØ PART 4: KEY INSIGHTS AND RECOMMENDATIONS")
print("-" * 50)

if df_clean is not None:
    print("üîç KEY INSIGHTS DISCOVERED:")
    
    # Publication trends
    if 'publish_year' in df_clean.columns:
        recent_years = df_clean['publish_year'] >= 2020
        covid_papers = df_clean[recent_years]
        pre_covid_papers = df_clean[~recent_years & (df_clean['publish_year'] >= 2015)]
        
        if len(covid_papers) > 0 and len(pre_covid_papers) > 0:
            growth_rate = ((len(covid_papers) / len(pre_covid_papers)) - 1) * 100
            print(f"   üìà COVID-19 research surge: {growth_rate:.1f}% increase in publication rate")
    
    # Journal concentration
    if 'journal' in df_clean.columns:
        top_5_journals = df_clean['journal'].value_counts().head(5).sum()
        concentration = (top_5_journals / len(df_clean)) * 100
        print(f"   üì∞ Research concentration: Top 5 journals publish {concentration:.1f}% of all papers")
    
    # Research focus
    if 'title' in df_clean.columns:
        clinical_terms = ['treatment', 'patient', 'clinical', 'therapy', 'drug']
        clinical_count = sum(df_clean['title'].str.contains('|'.join(clinical_terms), case=False, na=False))
        clinical_pct = (clinical_count / len(df_clean)) * 100
        print(f"   üè• Clinical focus: {clinical_pct:.1f}% of papers focus on clinical aspects")
    
    print(f"\nüí° RECOMMENDATIONS FOR FURTHER ANALYSIS:")
    print("   1. üî¨ Topic modeling to identify research clusters")
    print("   2. üìä Author collaboration network analysis") 
    print("   3. üåç Geographic distribution of research")
    print("   4. üìà Citation impact analysis")
    print("   5. ü§ñ Sentiment analysis of research abstracts")

print(f"\nüéâ ANALYSIS COMPLETE!")
print("=" * 50)
print("This notebook has successfully analyzed the CORD-19 dataset and provided")
print("insights into COVID-19 research publication patterns, journal distributions,")
print("and text characteristics. The cleaned dataset is ready for further analysis!")
print("=" * 50)
```
